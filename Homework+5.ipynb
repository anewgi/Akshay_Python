```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
pyber_data1 = pd.read_csv("/Users/anewgi/Downloads/ride_data.csv")
pyber_data2 = pd.read_csv("/Users/anewgi/Downloads/city_data.csv")
```

```python
avg_fare=pyber_data1.groupby(['city'])['fare'].mean().reset_index(name='Fare')
avg_fare
```

```python
total_rides = pyber_data1.groupby(['city']).size().reset_index(name='Ride_Count')
total_rides
```

```python
pyber_data=avg_fare.merge(total_rides,on='city').merge(pyber_data2, on='city')
pyber_data
```

```python
pyber_data.isnull().values.any()
```

```python
#Bubble Chart
fig, ax = plt.subplots()

color = np.where(pyber_data['type']=="Urban",'#75BBFD','-')
color[pyber_data['type']=="Suburban"]='#F5BF03'
color[pyber_data['type']=="Rural"] = '#FF7F50'
sc=ax.scatter(pyber_data.Ride_Count,pyber_data.Fare,s= pyber_data.driver_count*10,
              alpha=0.6, c=color,edgecolors='black',linewidths =2,)


fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 12
fig_size[1] = 9
plt.rcParams["figure.figsize"] = fig_size
fig.suptitle('Bubble Chart', fontsize=20)
plt.xlabel('Ride Count by City')
plt.ylabel('Average Fare')



ax.grid(True)



plt.show()
```

```python
percent_fare = ((pyber_data.groupby(['type'])['Fare'].sum())/pyber_data['Fare'].sum())*100
percent_fare
```

```python
pies = ["Rural",'Suburban','Urban']

colors = ["lightskyblue","lightcoral","gold"]
explode = (0.2,0,0)
```

```python

plt.pie(percent_fare, labels=pies, colors=colors, explode=explode,
        autopct="%1.1f%%", shadow=True, startangle=140)
```

```python
percent_fares = (pyber_data["type"].value_counts() / pyber_data['Fare'].sum())*100
percent_fares
```

```python
plt.axis("equal")
```

```python
plt.savefig("PyPies.png")
plt.show()
```

```python
percent_rides = ((pyber_data.groupby(['type'])['Ride_Count'].sum())/pyber_data['Ride_Count'].sum())*100
percent_rides
```

```python
plt.pie(percent_rides, labels=pies, colors=colors,explode=explode,
        autopct="%1.1f%%", shadow=True, startangle=360)

plt.axis("equal")

plt.savefig("PyPies.png")
plt.show()
```

```python
percent_drivers = ((pyber_data.groupby(['type'])['driver_count'].sum())/pyber_data['driver_count'].sum())*100
percent_drivers
```

```python
plt.pie(percent_drivers, labels=pies, colors=colors,explode=explode,
        autopct="%1.1f%%", shadow=False, startangle=360)

plt.axis("equal")

plt.savefig("PyPies.png")
plt.show()
```
